## Overriding Introduction

```python
'''
What Is Overriding?

When a child class defines a method with the same name as the parent class, the child version is used.

'''

class Animal:
  def __init__(self, name):
    self.name = name

  def make_noise(self):
    print("{} says, Grrrr".format(self.name))

pet1 = Animal("Rex")
pet1.make_noise() # Rex says, Grrrr



class Cat(Animal):

  def make_noise(self):
    print("{} says, Meow!".format(self.name))

'''
Here:

Cat inherits from Animal

It automatically gets:

- The constructor (__init__)

- The name attribute

But it overrides make_noise()

'''

pet2 = Cat("Maisy")
pet2.make_noise() # Maisy says, Meow!
```

```python
class Animal1:
  animal = "Janwar"

  def __init__(self, name):
    self.name = name

  def make_noise(self):
    print("{} says, Grrrr".format(self.name))

pet1 = Animal1("Rex")
pet1.make_noise() # Rex says, Grrrr



class Cat(Animal1):
  animal = "Akshay Kumar"

  def make_noise(self):
    print("{} says, Meow!".format(Cat.animal))

pet2 = Cat("Maisy")
print(pet2.animal)
pet2.make_noise() 
```

## super()

```python
'''
When overriding methods we sometimes want to still access the behavior of the parent method. 
In order to do that we need a way to call the method of the parent class. 
Python gives us a way to do that using 
super()
Preview: Docs Loading link description.
'''
class Animal:
    def __init__(self, name, sound = 'Grrrr'):
        self.name = name
        self.sound = sound
    def make_noise(self):
        print(f"{self.name} says {self.sound}")

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, sound = "Meow")
```

```python
pet_cat = Cat("Rachel")
pet_cat.make_noise()
```

# Multiple Inheritance: Part 2

```python
class Animal:
  def __init__(self, name):
    self.name = name

class Dog(Animal):
  def action(self):
    print("{} wags tail. Awwww".format(self.name))

class Wolf(Animal):
  def action(self):
    print("{} bites. OUCH!".format(self.name))

class Hybrid(Dog, Wolf):
  def action(self):
    super().action()
    Wolf.action(self)

my_pet = Hybrid("Fluffy")
my_pet.action() # Fluffy wags tail. Awwww
                # Fluffy bites. OUCH!

print(f"Method Resolution Order:\n",
      {Hybrid.__mro__}) 
```

# OOP Pillar: Abstraction

```python
from abc import ABC, abstractmethod

class AbstractEmployee(ABC):
  new_id = 1
  def __init__(self):
    self.id = AbstractEmployee.new_id
    AbstractEmployee.new_id += 1

  @abstractmethod
  def say_id(self):
    pass

# Write your code below
class Employee(AbstractEmployee):
  def say_id(self):
    return self.id

e1 = Employee()
print(e1.say_id())
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
        self._age = "?"
    
    def get_age(self):
        return self._age
    
    def set_age(self, new_age: int):
        if isinstance(new_age, int):
            self._age = new_age
        else:
            raise TypeError
        
    def delete_age(self):
        del self._age
        print("_age deleted")
        
```
