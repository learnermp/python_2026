# Lambda Function

```python
check_if_A_grade = lambda A_grade: 'Got an A!' if A_grade >= 90 else 'Did not get an A.'
```

```python
print(check_if_A_grade(91))
print(check_if_A_grade(70))
print(check_if_A_grade(20))
```

# Higher Order Function

Higher-order functions are possible because functions are first-class objects in Python, meaning that a function can be stored as a variable, passed as an argument to a function, returned by a function, and stored in data structures (lists, dictionaries, etc.).

```python
def add_tax(total):
    tax = total * 0.06
    new_total = total + tax
    return new_total

# total_bill(add_tax, 100)
```

```python
def add_tip(total):
  tip = total * .2
  new_total = total + tip
  return new_total
```

```python
def total_bill(func, value):
    total = func(value)
    return ("The total amount owned is $" + "{:.2f}".format(total) + "Thank you!:")
```

```python
print(total_bill(add_tax, 100))
print(total_bill(add_tip, 100))
```

```python
bills = [115, 120, 42]
new_bills = []

for i in range(len(bills)):
    total = add_tax(bills[i])
    new_bills.append("The total amount owned is $" + "{:.2f}".format(total) + "Thank you!:\n")
print(new_bills)
```

```python
def total_billing(func, list):
    new_bills = []
    for i in range(len(list)):
        total = func(list[i])
        new_bills.append("The total amount owned is $" + "{:.2f}".format(total) + "Thank you!:\n")
    return new_bills
```

```python
bills = [115, 120, 42]

bills_w_tax = total_billing(add_tax, bills)
print(bills_w_tax)
```

```python
bills = [115, 120, 42]
bills_w_tip = total_billing(add_tip, bills)
print(bills_w_tip)
```

The map() function in Python returns a map object, which is an iterator.

# Filter
Similar to map(), the filter() function takes a function and an iterable as arguments. Just as the name suggests, the goal of the filter() function is to “filter” values out of an iterable.

The filtering function should be a function that returns a boolean value: True or False. The returned filter object will hold only those elements of the passed iterable for which the filtering function returned True.

```python
names = ["margarita", "Linda", "Masako", "Maki", "Angela"]
M_names = filter(lambda name: name[0] == 'M' or name[0] == 'm', names)
print(list(M_names))
```

```python

l2 = [3, 5, 7]
result = filter(lambda x: x > 5, l2)
print(result)
print(list(result))
print(list(result))
```

# Reduce

```python
from functools import reduce
int_list = [1, 2, 3, 4]

reduce_int_list = reduce(lambda x, y: x * y, int_list)
print(reduce_int_list)
```

```python
from functools import reduce
letters = ['r', 'e', 'd', 'u', 'c', 'e']

reduced_word = reduce(lambda x, y: x + y, letters)
print(reduced_word)
```
