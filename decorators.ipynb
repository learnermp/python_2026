# Function within function

```python
def add_five(n):
    def add_two(n):
        return n + 2
    num_plus_two = add_two(n)
    print(num_plus_two)

add_five(10)
```

# Returning functions from functions

```python
def get_math_function(operation: str):
    def addition(n1, n2):
        return n1 + n2
    def subtraction(n1, n2):
        return n1 - n2

    if operation == '+':
        return addition
    elif operation == '-':
        return subtraction
```

```python
get_math_function('+')
```

```python
add_function = get_math_function('+')
print(add_function)
print(add_function(1, 2))
```

# Decorating a fucntion

```python
def title_decorator(print_name_function):
    def wrapper():
        print("Professor:")
        print_name_function()
    return wrapper

def print_my_name():
    print("Mrityunjay")

decorated_function = title_decorator(print_my_name)
decorated_function
```

```python
decorated_function()
```

```python
title_decorator(print_my_name)()
```
